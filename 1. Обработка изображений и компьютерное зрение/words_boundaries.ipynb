Работаю с NLP, поэтому просто было интересно что сможет такая простая реализация выделения границ

```python
import numpy as np
import imageio
import scipy.signal

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = imageio.imread('txt_photo.jpg')
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
plt.figure(figsize=(8, 8))
def photo_outlines_image(image_name):

  img = imageio.imread(str(image_name))
  temp_img = np.mean(img, axis=2)

  horizon = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])

  vertical = np.array([[-1,-1,-1],
                       [0, 0, 0],
                       [1, 1, 1]])
  h_img = scipy.signal.correlate2d(temp_img, horizon, mode='valid')
  v_img = scipy.signal.correlate2d(temp_img, vertical, mode='valid')

  plt.imshow(np.abs(h_img + v_img), cmap='gray')

  return plt.show()
```

```python
photo_outlines_image('txt_photo.jpg')
```
